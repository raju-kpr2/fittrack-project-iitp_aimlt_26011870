```python
#Create a pandas DataFrame using the employee data provided below
import pandas as pd

data = {
    "EmployeeID": [201,202,203,204,205,206,207,208,209,210],
    "Name": ["Sarah","Michael","Jennifer","David","Lisa","James","Emma","Robert","Sophia","William"],
    "Department": ["Engineering","Sales","Marketing","Engineering","HR","Sales","Engineering","Marketing","Sales","HR"],
    "Salary": [85000,62000,71000,92000,58000,68000,78000,75000,65000,60000],
    "Performance Score": [4,3,5,5,4,4,3,4,5,3],
    "Years of Service": [5,3,7,8,4,6,2,5,4,3]
}

employees_df = pd.DataFrame(data)
print(employees_df)
```

```python
#2 Select only Name, Department, Salary
employees_df[['Name','Department','Salary']]
```

```python
#3 Use iloc to display employee at row position 3
employees_df.iloc[3]
```

```python
#4 Use loc to display employee with EmployeeID 203
employees_df.set_index('EmployeeID').loc[203]
```

```python
#5 Employees from Engineering department
employees_df[employees_df['Department'] == 'Engineering']
```

```python
#6 Employees with Salary > 70000
employees_df[employees_df['Salary'] > 70000]
```

```python
#7 Sales employees with Performance Score ≥ 4
employees_df[
    (employees_df['Department'] == 'Sales') &
    (employees_df['Performance Score'] >= 4)
]
```

```python
# 8 Sort by Salary (Descending) and show top 3 highest paid employees
employees_df.sort_values('Salary', ascending=False).head(3)
```

```python
#9 Sort by Department then Salary (descending within each department)
employees_df.sort_values(
    ['Department','Salary'],
    ascending=[True, False]
)
```
