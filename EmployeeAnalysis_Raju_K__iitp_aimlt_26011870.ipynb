```python
## TASK 1 Python Code: Create employees.db with 40+ Records

import sqlite3
import random

# Connect to SQLite database (creates file if not exists)
conn = sqlite3.connect("employees.db")
cursor = conn.cursor()

# Create employees table
cursor.execute("""
CREATE TABLE IF NOT EXISTS employees (
    emp_id INTEGER PRIMARY KEY,
    name TEXT,
    department TEXT,
    salary REAL,
    years_experience INTEGER,
    performance_score REAL
)
""")

departments = ["Engineering", "Sales", "Marketing", "HR", "Finance"]

# Insert at least 40 records
for i in range(1, 41):
    name = f"Employee_{i}"
    dept = random.choice(departments)
    salary = random.randint(50000, 150000)
    experience = random.randint(1, 15)
    performance = round(random.uniform(1.0, 5.0), 2)

    cursor.execute("""
    INSERT INTO employees
    (emp_id, name, department, salary, years_experience, performance_score)
    VALUES (?, ?, ?, ?, ?, ?)
    """, (i, name, dept, salary, experience, performance))

# Save changes and close connection
conn.commit()
conn.close()

print("Database created and populated successfully.")
```

### Query 1 . Employees with performance_score ≥ 4.0 AND years_experience ≥ 3

SELECT name, department, salary, performance_score
FROM employees
WHERE performance_score >= 4.0
  AND years_experience >= 3
ORDER BY performance_score DESC
LIMIT 15;

### Query 2 . Salary BETWEEN 70,000 and 110,000 Department = Engineering OR Sales Sort by department, then salary DESC.

SELECT name, department, salary, performance_score
FROM employees
WHERE performance_score >= 4.0
  AND years_experience >= 3
ORDER BY performance_score DESC

### Query 3 . Count employees per department + average salary
Sort by avg_salary DESC.

SELECT department,
       COUNT(*) AS employee_count,
       AVG(salary) AS avg_salary
FROM employees
GROUP BY department
ORDER BY avg_salary DESC;





```python
#TASK 3
#1. Average Performance Score by Department (groupby)

import sqlite3
import pandas as pd

# Load data from SQLite
conn = sqlite3.connect("employees.db")
df = pd.read_sql("SELECT * FROM employees", conn)

# Group by department
avg_perf = df.groupby("department")["performance_score"].mean().reset_index()

print(avg_perf)
```

```python
#2. Replicate Query 1 Using ONLY Pandas
result = (
    df.loc[(df["performance_score"] >= 4.0) &
           (df["years_experience"] >= 3),
           ["name", "department", "salary", "performance_score"]]
      .sort_values(by="performance_score", ascending=False)
      .head(15)
)

print(result)
```

### ✅ SQL vs Pandas (150–200 words)

SQL and Pandas are both used to filter, sort, and analyze data, but they work in different ways. In SQL, data is stored in databases, and you use commands like **WHERE** to filter rows, **ORDER BY** to sort data, and **GROUP BY** with functions like AVG or COUNT to summarize results. In Pandas, which is a Python library, similar tasks are done using code such as boolean conditions with `.loc[]`, sorting with `.sort_values()`, and grouping with `.groupby()`.

SQL is best when data is very large and stored in a database. It can process huge datasets efficiently without loading everything into memory. Pandas works in memory, so it is better for smaller datasets or after you have already reduced the data size. It is especially useful for data cleaning, detailed analysis, and visualization in Python.

If the dataset is too large to fit into memory, SQL is usually the better choice because it runs on the database server and returns only the needed results. Pandas is best for deeper analysis once the data is manageable.
