```python
#1. Create Database Connection (Python + SQLite)

import sqlite3

# Create/connect to database
conn = sqlite3.connect("bookstore.db")

# Create cursor
cursor = conn.cursor()

print("Database created and connected successfully.")
```

#Sql script for table creation

## Books Table
CREATE TABLE IF NOT EXISTS Books (
    book_id INTEGER PRIMARY KEY AUTOINCREMENT,
    title TEXT NOT NULL,
    author TEXT NOT NULL,
    price REAL NOT NULL,
    stock_quantity INTEGER DEFAULT 0
);


## Books Customers
CREATE TABLE IF NOT EXISTS Customers (
    customer_id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    email TEXT UNIQUE NOT NULL,
    city TEXT,
    join_date TEXT
);

## Books Orders

CREATE TABLE IF NOT EXISTS Orders (
    order_id INTEGER PRIMARY KEY AUTOINCREMENT,
    customer_id INTEGER,
    book_id INTEGER,
    quantity INTEGER NOT NULL,
    order_date TEXT NOT NULL,
    total_amount REAL,
    FOREIGN KEY (customer_id) REFERENCES Customers(customer_id),
    FOREIGN KEY (book_id) REFERENCES Books(book_id)
);





```python
#Execute Table Creation from Python

cursor.execute("""CREATE TABLE IF NOT EXISTS Books (
    book_id INTEGER PRIMARY KEY AUTOINCREMENT,
    title TEXT NOT NULL,
    author TEXT NOT NULL,
    price REAL NOT NULL,
    stock_quantity INTEGER DEFAULT 0
)""")

cursor.execute("""CREATE TABLE IF NOT EXISTS Customers (
    customer_id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    email TEXT UNIQUE NOT NULL,
    city TEXT,
    join_date TEXT
)""")

cursor.execute("""CREATE TABLE IF NOT EXISTS Orders (
    order_id INTEGER PRIMARY KEY AUTOINCREMENT,
    customer_id INTEGER,
    book_id INTEGER,
    quantity INTEGER NOT NULL,
    order_date TEXT NOT NULL,
    total_amount REAL,
    FOREIGN KEY (customer_id) REFERENCES Customers(customer_id),
    FOREIGN KEY (book_id) REFERENCES Books(book_id)
)""")

conn.commit()

print("All tables created successfully.")
```

```python
# Function to show schema
def show_schema(table_name):
    print(f"\nSchema for {table_name}:")
    for row in cursor.execute(f"PRAGMA table_info({table_name})"):
        print(row)

show_schema("Books")
show_schema("Customers")
show_schema("Orders")
```

```python
#Task 3: Pandas Integration and Analysis
#Part A: SQL ➜ Pandas
#1. Load Tables into Pandas DataFrames

import sqlite3
import pandas as pd

# Connect to database
conn = sqlite3.connect("bookstore.db")

# Load tables
books_df = pd.read_sql("SELECT * FROM Books", conn)
customers_df = pd.read_sql("SELECT * FROM Customers", conn)
orders_df = pd.read_sql("SELECT * FROM Orders", conn)
```

```python
#2. Display First 3 Rows of Each DataFrame
print("Books:")
print(books_df.head(3))

print("\nCustomers:")
print(customers_df.head(3))

print("\nOrders:")
print(orders_df.head(3))
```

```python
#3. Comprehensive Order Report (Merge Tables)
report_df = orders_df.merge(customers_df, on="customer_id") \
                     .merge(books_df, on="book_id")

# Select required columns
report_df = report_df[[
    "order_id",
    "name",
    "city",
    "title",
    "quantity",
    "total_amount"
]]

# Rename columns
report_df.columns = [
    "Order ID",
    "Customer Name",
    "Customer City",
    "Book Title",
    "Quantity",
    "Total Amount"
]

print(report_df)
```

```python
#4. Calculations
avg_order_value = orders_df["total_amount"].mean()
print("Average Order Value:", avg_order_value)
orders_by_city = report_df.groupby("Customer City")["Order ID"].count()
print("\nTotal Orders by City:")
print(orders_by_city)
popular_book = report_df.groupby("Book Title")["Quantity"].sum().idxmax()
print("\nMost Popular Book:", popular_book)
```

```python
#Part B: Pandas ➜ SQL
#1. Create Discounts DataFrame

discounts_data = {
    "book_id": [1, 2, 3, 4, 5],
    "discount_percent": [10, 15, 5, 20, 12]
}

discounts_df = pd.DataFrame(discounts_data)

print(discounts_df)
```

```python
#Save DataFrame to SQL Table

discounts_df.to_sql("discounts", conn, if_exists="replace", index=False)

print("Discounts table created successfully.")

```

```python
#Verify Table Creation
verify_df = pd.read_sql("SELECT * FROM discounts", conn)
print(verify_df)
```

```python
#SQL Query — Books with Discounted Prices

discount_query = """
SELECT
    b.book_id,
    b.title,
    b.price,
    d.discount_percent,
    b.price * (1 - d.discount_percent / 100.0) AS discounted_price
FROM Books b
JOIN discounts d
ON b.book_id = d.book_id
"""

discounted_books_df = pd.read_sql(discount_query, conn)

print(discounted_books_df)
```
